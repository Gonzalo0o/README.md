#Parte 1

En un espectáculo de magia, un mago pide a un espectador que mezcle un conjunto de n
cartas. El espectador procede a tomar las cartas y las revuelve dentro del saco del mago,
sabiendo de antemano que almenos hay n y están enumeradas respectivamente. Con el
mago a espaldas de la mesa, coloca las cartas cara arriba en una fila. Diseñe un algoritmo
que ordene las cartas de menor a mayor de manera automática. De esta forma, una vez el
mago exponga su truco, estén ordenadas en fila. Puede realizarlo de la manera que mejor
le parezca usando los contenidos vistos en clases.

```python
print('Para este truco de magia te invitamos a indicar el número de carta que tu quieras, las veces que quieras. Para cuando no quieras ingresar más cartas apreta enter.')
cartas = [0]
while cartas[-1] != "":
  cartas.append(input("Ingrese una carta: "))
cartas.remove(0)
cartas.remove("")
cartas = list(map(int, cartas))
cartas.sort()
print('Las cartas ordenadas de forma ascendente son: ' + str(cartas))
```

## Pregunta 4
*Crear una columna de nombre Lesionados que tenga la totalidad de lesionados en un accidente, es decir, la suma de los accidentados "Graves", "MenosGrave" y "Leve"*

```python
n = df.shape[0]
df["Lesionados"] = 0
for i in range(n):
  df.loc[i, "Lesionados"] = df.loc[i, "Graves"] + df.loc[i, "MenosGrave"] + df.loc[i, "Leve"]
df.head(10)
```

## Pregunta 5

*Utilizando Python responda las siguientes preguntas ¿En qué esquina de Santiago hay más accidentados? ¿En qué comuna de santiago hay más accidentes de transito? ¿En qué esquina de Santiago hay más fallecidos?*

```python
df_lesionados = (df.sort_values("Lesionados", ascending=False))
print("La esquina de Santiago donde hay más accidentados es", df_lesionados.iloc[0,2].split(sep=", ")[0])
```

```python
df_accidentes = (df.sort_values("Accidentes", ascending=False))
print("La comuna de Santiago donde hay más accidentes de transito es", df_accidentes.iloc[0,2].split(sep=", ")[1])
```

```python
df_fallecidos = (df.sort_values("Fallecidos", ascending=False))
print("La esquina de Santiago donde hay más fallecidos es", df_fallecidos.iloc[0,2].split(sep=", ")[0])
```
